## Разведочный анализ данных (EDA)

**Датасет:** Wine (Scikit‑learn)

Датасет содержит химические характеристики вина трёх сортов.

## 1. Загрузка и описание датасета

```python
from sklearn.datasets import load_wine
import pandas as pd

wine = load_wine()
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df['target'] = wine.target
df.head()
```

## 2. Основная информация о датасете

```python
df.info()
```

```python
df.describe()
```

## 3. Распределение классов

```python
df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt

df['target'].value_counts().plot(kind='bar')
plt.title('Распределение классов')
plt.show()
```

## 4. Гистограммы признаков

```python
df.hist(figsize=(12, 10))
plt.tight_layout()
plt.show()
```

## 5. Boxplot каждого признака

```python
df.iloc[:, :-1].plot(kind='box', figsize=(14,6))
plt.xticks(rotation=90)
plt.title('Boxplot признаков')
plt.show()
```

## 6. Парные диаграммы (Scatter Matrix)

```python
pd.plotting.scatter_matrix(df, figsize=(14, 14))
plt.show()
```

## 7. Корреляционная матрица

```python
corr = df.corr()
corr
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
plt.imshow(corr, cmap='coolwarm')
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns, rotation=90)
plt.yticks(range(len(corr)), corr.columns)
plt.title('Корреляционная матрица (heatmap)')
plt.show()
```
